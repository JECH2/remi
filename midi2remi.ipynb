# Convert to REMI events

## 1. Read midi into "Item"

```python
note_items = midi2mumidi.read_items(filelist[0])
```

```python
print(*note_items, sep='\n')
```

```python
grids = np.arange(0, note_items[-1].start, 689, dtype=int)
print(len(grids))
```

## 2. Quantize note items

```python
note_items = midi2mumidi.quantize_items(note_items)
```

```python
print(*note_items, sep='\n')
```

## 3. extract chord (if needed)

```python
chord_items = utils.extract_chords(note_items)
```

```python
print(*chord_items, sep='\n')
```

## 4. group items

```python
items = note_items
#items = chord_items + tempo_items + note_items
max_time = note_items[-1].end
print(max_time)
groups = midi2mumidi.group_items(items, max_time)
```

```python
for g in groups:
    print(*g, sep='\n')
    print()
```

## 5. "Item" to "Event"

```python
events = midi2mumidi.item2event(groups)
```

```python
print(*events[:30], sep='\n')
```
